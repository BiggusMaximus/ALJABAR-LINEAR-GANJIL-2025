```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Rectangle
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
plt.rcParams.update({
    'font.size': 8,
    'text.usetex': True,
    'text.latex.preamble': r'\usepackage{amsfonts}\usepackage{amsmath}',
    "font.family": "Times New Roman"
})
```

# **2D**

## Parametric Equation pada Garis

```python
fig, ax = plt.subplots(figsize=(6,6), dpi=200) 
ax.set_xlim(-1, 10)
ax.set_ylim(-1, 10)
ax.set_aspect('equal')

# ----------------------------
# Line through x0 parallel to v
# ----------------------------
v = np.array([3, 1])   # direction vector
x_0 = np.array([3, 2])
t_vals = np.linspace(-2, 2, 100)  # parameter range

# Line points: x = x0 + tv
line_points = np.array([x_0[0] + t_vals * v[0],
                        x_0[1] + t_vals * v[1]])

# Draw the line
ax.plot(line_points[0], line_points[1], color="gray", linestyle="--", label=r"L: $x=x_0+tv$")

# Mark x0
ax.scatter(*x_0, color="black", zorder=5)
ax.text(x_0[0]+0.2, x_0[1]-0.3, r"$x_0$", fontsize=12)

# Mark x
ax.scatter(*v, color="black", zorder=5)
ax.text(x_0[0] + v[0]+0.2, x_0[1] + v[1]-0.3, r"$x$", fontsize=12)

# Draw vector v from origin
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color="blue")
ax.text(v[0]/2-0.5, v[1]/2+0.25, r"$\vec{v}$", fontsize=12, color="blue")

# Draw same vector v starting at x0
ax.quiver(x_0[0], x_0[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color="red")
ax.text(x_0[0]+v[0]/2-0.5, x_0[1]+v[1]/2+0.5, r"$x-x_0$", fontsize=12, color="red")

# Axes
ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)
ax.legend(fontsize=14)

plt.show()
```

## Translasi antar garis

```python
fig, ax = plt.subplots(figsize=(6,6), dpi=200) 
ax.set_xlim(-1, 10)
ax.set_ylim(-1, 10)
ax.set_aspect('equal')

# ----------------------------
# Line through x0 parallel to v
# ----------------------------
v = np.array([3, 1])   # direction vector
x_0 = np.array([3, 5])
t_vals = np.linspace(-10, 10, 100)  # parameter range

# Line points: x = x0 + tv
line_points = np.array([x_0[0] + t_vals * v[0],
                        x_0[1] + t_vals * v[1]])

# Draw the line
ax.plot(line_points[0], line_points[1], color="b", linestyle="--", label=r"L: $x=x_0+tv$", zorder=-1)

# Mark x0
ax.scatter(*x_0, color="black", zorder=5)
ax.text(x_0[0]+0.2, x_0[1]-0.3, r"$x_0$", fontsize=12)

# Mark x
ax.scatter(*v, color="black", zorder=5)
ax.text(x_0[0] + v[0]+0.2, x_0[1] + v[1]-0.3, r"$x$", fontsize=12)


# Line points: x = x0 + tv
line_points = np.array([t_vals * v[0],
                        t_vals * v[1]])

# Draw the line
ax.plot(line_points[0], line_points[1], color="r", linestyle="--", label=r"L: $x=tv$", zorder=-1)

# Draw the x_0 quiver through origin
ax.quiver(0, 0, x_0[0], x_0[1], angles='xy', scale_units='xy', scale=1, color="k")

# Draw vector v from origin
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color="k")
ax.text(v[0]/2-0.5, v[1]/2+0.25, r"$\vec{v}$", fontsize=12, color="k")

# Draw same vector v starting at x0
ax.quiver(x_0[0], x_0[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color="k")
ax.text(x_0[0]+v[0]/2-0.5, x_0[1]+v[1]/2+0.5, r"$x-x_0$", fontsize=12, color="k")

# Axes
ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)
ax.legend(fontsize=14)

plt.show()
```

# **3D**

## Parametric Equation pada Garis

```python
x0 = np.array([1, 2, 0])
v = np.array([2, -1, 3])

# Parameter range
t_vals = np.linspace(-2, 2, 100)
X = x0[0] + v[0] * t_vals
Y = x0[1] + v[1] * t_vals
Z = x0[2] + v[2] * t_vals

# Plot
fig = plt.figure(figsize=(8,6), dpi=150)
ax = fig.add_subplot(111, projection='3d')

# Line
ax.plot(X, Y, Z, color="gray", label=r"$L:\mathbf{x}(t)=\mathbf{x}_0+t\mathbf{v}$")

# Starting point x0
ax.scatter(*x0, color="red", s=50, label=r"$\mathbf{x}_0$")

# Direction vector
ax.quiver(*x0, *v, color="blue", arrow_length_ratio=0.2, label=r"$\vec{v}$")

# Labels
anchor = 7
ax.plot([-anchor, anchor], [0,0], [0,0], color="black", linewidth=0.8)  # X-axis
ax.plot([0,0], [-anchor, anchor], [0,0], color="black", linewidth=0.8)  # Y-axis
ax.plot([0,0], [0,0], [-anchor, anchor], color="black", linewidth=0.8)  # Z-axis
ax.grid(False)

ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.set_zlim(-1, 5)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend(fontsize=14)

plt.show()
```

## Parametric Equation pada Bidang

```python
x0 = np.array([1, 2, 0])     # point on the plane
v1 = np.array([1, 0, 2])     # direction vector 1
v2 = np.array([0, 2, 1])     # direction vector 2

# ----------------------------
# Create a grid of parameters s, t
# ----------------------------
s_vals = np.linspace(-2, 2, 20)
t_vals = np.linspace(-2, 2, 20)
S, T = np.meshgrid(s_vals, t_vals)

# Parametric equation of the plane: X = x0 + s v1 + t v2
X = x0[0] + S * v1[0] + T * v2[0]
Y = x0[1] + S * v1[1] + T * v2[1]
Z = x0[2] + S * v1[2] + T * v2[2]

# ----------------------------
# Plot
# ----------------------------
fig = plt.figure(figsize=(8,6), dpi=150)
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(X, Y, Z, alpha=0.4, color="lightblue")

# Plot the point x0
ax.scatter(*x0, color="red", s=50, label=r"$\vec{x}_0$")

# Plot the direction vectors from x0
ax.quiver(*x0, *v1, color="blue" , arrow_length_ratio=0.2, label=r"$\vec{v}_1$")
ax.text(
    (v1[0] + x0[0])/2, 
    (v1[1] + x0[1])/2, 
    (v1[2] + x0[2])/2, r"$\vec{v}_1$", fontsize=12, color="blue")

ax.quiver(*x0, *v2, color="green", arrow_length_ratio=0.2, label=r"$\vec{v}_2$")
ax.text(
    (v2[0] + x0[0])/2, 
    (v2[1] + x0[1])/2, 
    (v2[2] + x0[2])/2, r"$\vec{v}_2$", fontsize=12, color="green")

x = x0 + v1 + v2
ax.quiver(*x0, *x, color="r", arrow_length_ratio=0.2, label=r"$x$")
ax.text(
    (x[0] + x0[0]), 
    (x[1] + x0[1]), 
    (x[2] + x0[2]), r"$x$", fontsize=12, color="r")

anchor = 7
ax.plot([-anchor, anchor], [0,0], [0,0], color="black", linewidth=0.8)  # X-axis
ax.plot([0,0], [-anchor, anchor], [0,0], color="black", linewidth=0.8)  # Y-axis
ax.plot([0,0], [0,0], [-anchor, anchor], color="black", linewidth=0.8)  # Z-axis

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.grid(False)
ax.legend(fontsize=12)
plt.show()
```

## Translasi antar bidang

```python

x0 = np.array([10, 5, 30])     # point on the plane
v1 = np.array([5, 0, 5])     # direction vector 1
v2 = np.array([0, 5, 5])     # direction vector 2

# ----------------------------
# Create a grid of parameters s, t
# ----------------------------
s_vals = np.linspace(-2, 2, 20)
t_vals = np.linspace(-2, 2, 20)
S, T = np.meshgrid(s_vals, t_vals)

# Parametric equation of the plane: X = x0 + s v1 + t v2
X = x0[0] + S * v1[0] + T * v2[0]
Y = x0[1] + S * v1[1] + T * v2[1]
Z = x0[2] + S * v1[2] + T * v2[2]

# ----------------------------
# Plot
# ----------------------------
fig = plt.figure(figsize=(8,6), dpi=150)
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(X, Y, Z, alpha=0.4, color="lightblue", label=r"$x = x_0 + t_1v_1 + t_2v_2$")
ax.text(
    np.max(X), 
    np.max(Y), 
    np.max(Z), r"$x = x_0 + t_1v_1 + t_2v_2$", fontsize=12, color="k")

# Plot the point x0
ax.scatter(*x0, color="red", s=50, label=r"$\vec{x}_0$")

# Plot the direction vectors from x0


x = x0 + v1 + v2
ax.quiver(*np.array([0, 0, 0]), *x0, color="r", arrow_length_ratio=0.2)



# Parametric equation of the plane: X = x0 + s v1 + t v2
X1 = S * v1[0] + T * v2[0]
Y1 = S * v1[1] + T * v2[1]
Z1 = S * v1[2] + T * v2[2]
x0 = np.array([0, 0, 0])

# Plot the plane
ax.plot_surface(X1, Y1, Z1, alpha=0.4, color="maroon", label=r"$x = t_1v_1 + t_2v_2$")
ax.text(
    np.max(X1), 
    np.max(Y1), 
    np.max(Z1), r"$x = t_1v_1 + t_2v_2$", fontsize=12, color="k")

x = x0 + v1 + v2
ax.quiver(*x0, *x, color="r", arrow_length_ratio=0.2)


ax.quiver(*x0, *v1, color="blue" , arrow_length_ratio=0.2)
ax.text(
    (v1[0] + x0[0])/2, 
    (v1[1] + x0[1])/2, 
    (v1[2] + x0[2])/2, r"$\vec{v}_1$", fontsize=12, color="blue")

ax.quiver(*x0, *v2, color="green", arrow_length_ratio=0.2)
ax.text(
    (v2[0] + x0[0])/2, 
    (v2[1] + x0[1])/2, 
    (v2[2] + x0[2])/2, r"$\vec{v}_2$", fontsize=12, color="green")

anchor = np.max(Z1)
ax.plot([-anchor, anchor], [0,0], [0,0], color="black", linewidth=0.8)  # X-axis
ax.plot([0,0], [-anchor, anchor], [0,0], color="black", linewidth=0.8)  # Y-axis
ax.plot([0,0], [0,0], [-anchor, anchor], color="black", linewidth=0.8)  # Z-axis



ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.grid(False)
ax.legend(fontsize=12)
plt.show()
```

## **Cross Product**

```python
a = np.array([1, 2, 0])
b = np.array([2, -1, 3])
c = np.cross(a, b)  # Cross product

# Plot
fig = plt.figure(figsize=(8,6), dpi=150)
ax = fig.add_subplot(111, projection='3d')

# Origin
origin = np.array([0,0,0])

# Vectors
ax.quiver(*origin, *a, color="blue", arrow_length_ratio=0.15, label=r"$\vec{a}$")
ax.quiver(*origin, *b, color="green", arrow_length_ratio=0.15, label=r"$\vec{b}$")
ax.quiver(*origin, *c, color="red", arrow_length_ratio=0.15, label=r"$\vec{a}\times\vec{b}$")

# Axes lines
anchor = 10

ax.plot([-anchor, anchor], [0,0], [0,0], color="black", linewidth=0.8)  # X-axis
ax.plot([0,0], [-anchor, anchor], [0,0], color="black", linewidth=0.8)  # Y-axis
ax.plot([0,0], [0,0], [-anchor, anchor], color="black", linewidth=0.8)  # Z-axis
ax.grid(False)

# Limits
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)

# Labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend(fontsize=14)

plt.show()
```

```python

# Vectors
a = np.array([1, 2, 0])
b = np.array([2, -1, 3])
c = np.cross(a, b)  # Cross product

# Area of parallelogram = |a  b|
area = np.linalg.norm(c)

# Plot
fig = plt.figure(figsize=(8,6), dpi=150)
ax = fig.add_subplot(111, projection='3d')

# Origin
origin = np.array([0,0,0])

# Vectors
ax.quiver(*origin, *a, color="blue", arrow_length_ratio=0.15, label=r"$\vec{a}$")
ax.quiver(*origin, *b, color="green", arrow_length_ratio=0.15, label=r"$\vec{b}$")
ax.quiver(*origin, *c, color="red", arrow_length_ratio=0.15, label=r"$\vec{a}\times\vec{b}$")

# Parallelogram spanned by a and b
parallelogram = np.array([
    origin, 
    a, 
    a + b, 
    b
])
ax.add_collection3d(
    Poly3DCollection([parallelogram], alpha=0.3, facecolor="cyan")
)

# Axes lines
anchor = 10
ax.plot([-anchor, anchor], [0,0], [0,0], color="black", linewidth=0.8)  # X-axis
ax.plot([0,0], [-anchor, anchor], [0,0], color="black", linewidth=0.8)  # Y-axis
ax.plot([0,0], [0,0], [-anchor, anchor], color="black", linewidth=0.8)  # Z-axis
ax.grid(False)

# Limits
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)

# Labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend(fontsize=14)

# Title with parallelogram area
ax.set_title(f"Area of parallelogram = {area:.2f}", fontsize=14)

plt.show()
```
