```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Rectangle
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
plt.rcParams.update({
    'font.size': 8,
    'text.usetex': True,
    'text.latex.preamble': r'\usepackage{amsfonts}\usepackage{amsmath}',
    "font.family": "Times New Roman"
})
```

```python
fig, ax = plt.subplots(figsize=(6,6), dpi=200)
ax.set_xlim(-1, 10)
ax.set_ylim(-1, 10)
ax.set_aspect('equal')

# Define vectors
u = np.array([3, 2])
v = np.array([2, 5])

# Draw u
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color="blue")
ax.text(u[0]/2, u[1]/2-0.3, r"$\vec{u}$", fontsize=12, color="blue")

# Draw v
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color="green")
ax.text(v[0]/2, v[1]/2+0.2, r"$\vec{v}$", fontsize=12, color="green")

# Draw u+v
uv = u + v
ax.quiver(0, 0, uv[0], uv[1], angles='xy', scale_units='xy', scale=1, color="red")
ax.text(uv[0]/2, uv[1]/2+0.2, r"$\vec{u}+\vec{v}$", fontsize=12, color="red")

# Axes
ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)
plt.show()
```

```python
# ----------------------------
# Example 2: General vector space (polynomials as vectors)
# ----------------------------
fig, ax = plt.subplots(figsize=(6,6), dpi=200)

# Define polynomials as functions
x = np.linspace(-2*np.pi, 2*np.pi, 50)
p = np.sin(x)

ax.plot(x, p , label=r"$y(x) = \sin(x)$")
ax.stem(x, p )
# Axes
ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)
ax.legend(fontsize=12)

plt.show()
```
