```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Rectangle
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from ipywidgets import interact, FloatSlider
import matplotlib.patches as mpatches
plt.rcParams.update({
    'font.size': 8,
    'text.usetex': True,
    'text.latex.preamble': r'\usepackage{amsfonts}\usepackage{amsmath}',
    "font.family": "Times New Roman"
})
```

```python

A = np.array([[3, 0],
              [8, -1]])
x = np.array([1, 2])

# Compute Ax
Ax = A @ x

# Set up the plot
plt.figure(figsize=(6,6), dpi=150)
origin = np.array([[0, 0],[0, 0]])  # origin for quiver

# Plot the transformed vector Ax in red
plt.quiver(*origin[:,0], Ax[0], Ax[1], color='red', scale=1, scale_units='xy', angles='xy', label='$Ax = \lambda x \; (\lambda=3)$')
# Plot the original vector x in blue
plt.quiver(*origin[:,0], x[0], x[1], color='blue', scale=1, scale_units='xy', angles='xy', label='x')


# Set plot limits
all_vectors = np.array([x, Ax])
max_val = np.max(np.abs(all_vectors)) + 1
plt.xlim(-max_val, max_val)
plt.ylim(-max_val, max_val)

# Add grid, legend, and labels
plt.grid(False)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
