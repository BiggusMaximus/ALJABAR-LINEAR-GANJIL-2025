# **Vektor 2D - $\mathbb{R}^2$**

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6,6), dpi=250)
ax.set_xlim(-2, 6)
ax.set_ylim(-5, 5)
ax.set_aspect('equal')

# Points and vectors
p0 = np.array([0, 0])     # initial point for v1
v1 = np.array([2, 3])     # vector v1

p1 = np.array([-1, -1])   # initial point for v2
v2 = v1.copy()            # make v2 equivalent to v1

# Mark and label p0
ax.scatter([p0[0]], [p0[1]], color='r', marker='o', zorder=5)
ax.text(p0[0] + 0.15, p0[1] - 0.5, 
        r"$\vec{p}_0 = \begin{bmatrix} 0\\0 \end{bmatrix} \rightarrow \substack{Initial\\Point}$", 
        fontsize=12)

# Terminal point of v1
terminal_v1 = p0 + v1
ax.scatter(terminal_v1[0], terminal_v1[1], color='r', marker='o', zorder=5)
ax.text(terminal_v1[0] + 0.15, terminal_v1[1], 
        r"$\vec{p}_1 = \begin{bmatrix} 2\\3 \end{bmatrix} \rightarrow \substack{Terminal\\Point}$", 
        fontsize=12)

# Draw v1 from p0
ax.quiver(p0[0], p0[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
          color="blue", label=r"$\vec{v}_1$")
ax.text(v1[0]/2, v1[1]/2 - 0.25, 
        fr"$\vec{{v}}_1 = \begin{{bmatrix}} {v1[0]}\\{v1[1]} \end{{bmatrix}}$", 
        fontsize=12)

# Draw v2 from p1 (equivalent to v1)
terminal_v2 = p1 + v2
ax.quiver(p1[0], p1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
          color="green", label=r"$\vec{v}_2$")
ax.text(p1[0] + v2[0]/2 - 0.5, p1[1] + v2[1]/2 + 0.5, 
        fr"$\vec{{v}}_2$", 
        fontsize=12)

# Axes
ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)

plt.show()
```

## **Penjumlahan Vektor**

```python
fig, ax = plt.subplots(figsize=(6,6), dpi=200)
ax.set_xlim(-3, 6)
ax.set_ylim(-3, 6)
ax.set_aspect('equal')

# Define vectors
v1 = np.array([2, 1])
v2 = np.array([1, 3])
v3 = v1 + v2  # vector sum
v4 = v1 - v2  # vector sum

# Plot vector a
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color="blue")
ax.text(v1[0]/2, v1[1]/2 + 0.35, r"$\vec{v_1}$", fontsize=12, color="blue")

# Plot vector -a
ax.quiver(0, 0, -v1[0], -v1[1], angles='xy', scale_units='xy', scale=1, color="blue")
ax.text(-v1[0]/2, -v1[1]/2 - 0.25, r"$-\vec{v_1}$", fontsize=12, color="blue")

# Plot vector 2a
ax.quiver(0, 0, 2*v1[0], 2*v1[1], angles='xy', scale_units='xy', scale=1, color="blue")
ax.text(2*v1[0]/2, 2*v1[1]/2 + 0.35, r"$2\vec{v_1}$", fontsize=12, color="blue")

# Plot vector b
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color="green")
ax.text(v2[0]/2 - 0.4, v2[1]/2, r"$\vec{v_2}$", fontsize=12, color="green")

# Plot vector -b
ax.quiver(0, 0, -v2[0], -v2[1], angles='xy', scale_units='xy', scale=1, color="green")
ax.text(-v2[0]/2 - 0.7, -v2[1]/2, r"$-\vec{v_2}$", fontsize=12, color="green")

# Plot resultant vector (a+b)
ax.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color="red")
ax.text(v3[0]/2 + 0.25, v3[1]/2 + 0.2, r"$\vec{v_1} + \vec{v_2}$", fontsize=12, color="red")

# Plot resultant vector (a+b)
ax.quiver(0, 0, v4[0], v4[1], angles='xy', scale_units='xy', scale=1, color="red")
ax.text(v4[0]/2 + 0.25, v4[1]/2 + 0.2, r"$\vec{v_1} - \vec{v_2}$", fontsize=12, color="red")

# Show "parallelogram law"
ax.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
          color="gray", alpha=0.5, linestyle="dashed")
ax.quiver(v2[0], v2[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
          color="gray", alpha=0.5, linestyle="dashed")

# Show "parallelogram law"
ax.quiver(v1[0], v1[1], -v2[0], -v2[1], angles='xy', scale_units='xy', scale=1, 
          color="k", alpha=0.5, linestyle="dashed")
ax.quiver(-v2[0], -v2[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
          color="k", alpha=0.5, linestyle="dashed")

# Axes
ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)

plt.show()
```

## **3 atau lebih Vektor**

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6,6), dpi=200)
ax.set_xlim(-1, 8)
ax.set_ylim(-1, 8)
ax.set_aspect('equal')

# Define three vectors
v1 = np.array([2, 1])
v2 = np.array([1, 3])
v3 = np.array([3, 2])

# Sum of all three
v_sum = v1 + v2 + v3

# --- Plot v1 from origin
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color="blue")
ax.text(v1[0]/2, v1[1]/2+0.2, r"$\vec{v}_1$", fontsize=12, color="blue")

# --- Plot v2 starting at head of v1
ax.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color="green")
ax.text(v1[0] + v2[0]/2, v1[1] + v2[1]/2-0.5, r"$\vec{v}_2$", fontsize=12, color="green")

# --- Plot v3 starting at head of v1+v2
v1v2 = v1 + v2
ax.quiver(v1v2[0], v1v2[1], v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color="orange")
ax.text(v1v2[0] + v3[0]/2, v1v2[1] + v3[1]/2+0.2, r"$\vec{v}_3$", fontsize=12, color="orange")

# --- Resultant vector from origin to final point
ax.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color="red", linewidth=2)
ax.text(v_sum[0]/2+0.2, v_sum[1]/2, r"$\vec{v}_1 + \vec{v}_2 + \vec{v}_3$", fontsize=12, color="red")

# Axes
ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)

plt.show()
```

## **Paralellogram (1)**

```python
fig, ax = plt.subplots(figsize=(6,6), dpi=200)
ax.set_xlim(-0.5, 5)
ax.set_ylim(-0.5, 5)
ax.set_aspect('equal')

# ----------------------------
# Example: Parallelogram
# ----------------------------
u = np.array([2, 0])   # first vector
v = np.array([1, 3])   # second vector

# Draw u
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color="blue")
ax.text(u[0]/2, u[1]/2 - 0.25, r"$\vec{u}$", fontsize=14, color="blue")

# Draw v
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color="red")
ax.text(v[0]/2 - 0.25, v[1]/2 - 0.25, r"$\vec{v}$", fontsize=14, color="red")

# Draw u shifted by v
ax.quiver(v[0], v[1], u[0], u[1], angles='xy', scale_units='xy', scale=1, color="blue", alpha=0.6, linestyle="--")

# Draw v shifted by u
ax.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color="red", alpha=0.6, linestyle="--")

# u + v
ax.quiver(0, 0, u[0] + v[0], u[1] + v[1], angles='xy', scale_units='xy', scale=1, color="black", linestyle="--")
ax.text((u[0] + v[0])/2 + 0.25, (u[1] + v[1])/2, r"$\vec{u} + \vec{v}$", fontsize=14, color="k")

# u - v
ax.quiver(v[0], v[1], u[0] - v[0], u[1] - v[1], angles='xy', scale_units='xy', scale=1, color="cyan", linestyle="--")
ax.text((u[0] + v[0])/2 - 0.45, (u[1] + v[1])/2 - 0.65, r"$\vec{u} - \vec{v}$", fontsize=14, color="k")

# Coordinates of parallelogram
points = np.array([[0,0], u, u+v, v])

# Draw parallelogram
ax.fill(points[:,0], points[:,1], alpha=0.2, color="blue", edgecolor="black")

# Axes
ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)

plt.show()
```

## **Orthogonality**

```python
import numpy as np
import matplotlib.pyplot as plt

P0 = np.array([2, -1, 4])           # point outside plane
n = np.array([1, 2, 3])             # normal vector

# --- Define the plane ax+by+cz = d
d = np.dot(n, P0)                   # plane constant

xx, yy = np.meshgrid(np.linspace(-2, 6, 10), np.linspace(-2, 6, 10))
zz = (d - n[0]*xx - n[1]*yy) / n[2]

# --- Foot of perpendicular (projection of P0 onto plane)
t = (d - np.dot(n, P0)) / np.dot(n, n)   # scalar along n
P = P0 + t*n                              # projected point on plane

# --- Distance formula
D = np.abs(np.dot(n, P0) - d) / np.linalg.norm(n)

# 3D plot
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_aspect('equal')

# --- Draw the plane
ax.plot_surface(xx, yy, zz, alpha=0.3, color="green")

# --- Normal vector at P0
ax.quiver(P0[0], P0[1], P0[2], n[0], n[1], n[2],
          color="blue", arrow_length_ratio=0.1, linewidth=2)
ax.text(P0[0]+n[0]/2, P0[1]+n[1]/2, P0[2]+n[2]/2,
        rf"$\vec{{n}} = \begin{{bmatrix}}{n[0]} \\ {n[1]} \\ {n[2]}\end{{bmatrix}}$",
        fontsize=12, color="blue")

# --- Mark point P0
ax.scatter(*P0, color="red", s=50)
ax.text(P0[0]+0.2, P0[1]-0.3, P0[2], r"$P_0(2,-1,4)$", fontsize=12, color="red")

# --- Mark projection point P
ax.scatter(*P, color="purple", s=50)
ax.text(P[0]+0.2, P[1]+0.2, P[2], r"$P$", fontsize=12, color="purple")

# --- Draw perpendicular line from P0 to plane
ax.plot([P0[0], P[0]], [P0[1], P[1]], [P0[2], P[2]], 
        color="black", linewidth=2, linestyle="--")

# --- Annotate distance
ax.text((P0[0]+P[0])/2, (P0[1]+P[1])/2, (P0[2]+P[2])/2 + 0.2,
        rf"$D={D:.2f}$", fontsize=12, color="black")

# Axes setup
ax.set_xlim([-2, 6])
ax.set_ylim([-2, 6])
ax.set_zlim([-2, 8])
ax.set_xlabel("X", fontsize=12)
ax.set_ylabel("Y", fontsize=12)
ax.set_zlabel("Z", fontsize=12)

# Draw Cartesian axes
anchor = 7
ax.plot([-anchor, anchor], [0,0], [0,0], color="black", linewidth=0.8)  # X-axis
ax.plot([0,0], [-anchor, anchor], [0,0], color="black", linewidth=0.8)  # Y-axis
ax.plot([0,0], [0,0], [-anchor, anchor], color="black", linewidth=0.8)  # Z-axis
ax.grid(False)

plt.show()
```
